```python
# Apply Adaptive Gamma Correlation on Images
# Import Library
import glob as gb
import os
import cv2     
import numpy as np

# Read The Data For Training set 
dataPath = 'D:/AffectNet/'
outputPath = 'D:/AffectNet_Gamma/result/'

# Loop through each folder in the training set
for folder in os.listdir(dataPath + 'train_set'):
    # Create the folder in the output path if it doesn't exist
    result_folder_path = os.path.join(outputPath, folder)
    if not os.path.exists(result_folder_path):
        os.makedirs(result_folder_path)
    
    # Get all image files in the current folder
    files = gb.glob(pathname=str(dataPath + 'ang//' + folder + '/*.*g'))
    
    # Loop through each file
    for file in files:
        # Load image
        img = cv2.imread(file)
        if img is None:
            raise ValueError("Image not found or invalid path.")

        # Convert to float for processing
        img_float = img.astype(np.float32) / 255.0

        # Compute mean luminance
        mean_luminance = np.mean(img_float)

        # Compute adaptive gamma (range 0.6–1.4)
        if mean_luminance < 0.5:
            gamma = 0.3 + (mean_luminance / 0.5) * (1.0 - 0.6)
        else:
            gamma = 1.0 + ((mean_luminance - 0.5) / 0.5) * (1.4 - 1.0)

        # Apply gamma correction
        corrected = np.power(img_float, gamma)
        corrected_uint8 = np.clip(corrected * 255, 0, 255).astype(np.uint8)

        # Convert back to BGR color space
        # enhanced_image = cv2.cvtColor(enhanced_image, cv2.COLOR_LAB2BGR)
        
        # Get the file name without the directory path
        filename = os.path.basename(file)
        
        # Save the enhanced image to the corresponding result folder
        save_path = os.path.join(result_folder_path, filename)
        cv2.imwrite(save_path, corrected_uint8)
```
